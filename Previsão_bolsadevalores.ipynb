Previsão da ação

```python
#importando as bibliotecas
from sklearn.feature_selection import SelectKBest
from sklearn.model_selection import GridSearchCV
from sklearn.neural_network import MLPRegressor
from sklearn.preprocessing import MinMaxScaler
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

```

```python
#criando novos campos de médias moveis
df_cotacoes['media5d']= df_cotacoes['Adj Close'].rolling(5).mean()
```

```python
df_cotacoes
```

```python
#Ajustando o fechamento um dia para frente
df_cotacoes['Adj Close']= df_cotacoes['Adj Close'].shift(-1)
df_cotacoes.head(10)
```

```python
#retirando todos os dados nulos
df_cotacoes.dropna(inplace=True)
df_cotacoes
```

```python
#Verificando quantidade de linhas
qtd_linhas = len(df_cotacoes)
print(qtd_linhas)
#70% treinamento
qtd_linha_treino = qtd_linhas - 30
#20% teste
qtd_linha_teste = qtd_linhas - 10
#10% validação
qtd_linha_validacao = qtd_linha_treino - qtd_linha_teste

info = {
    f"linhas treino: {qtd_linha_treino} "
    f"linhas teste: {qtd_linha_treino}:{qtd_linha_teste} "
    f"linhas validacao: {qtd_linha_teste}:{qtd_linhas} "
}

info

```

```python
#reindexando
# df_cotacoes = df_cotacoes.reset_index(inplace=True)
df_cotacoes = df_cotacoes.reset_index(drop=True)
print(df_cotacoes)
```

```python
#Separação de features e labels
features = df_cotacoes.drop(['Adj Close'], axis=1)
label = df_cotacoes['Adj Close']
```

```python
#Escolhendo as melhores features
feature_list = ('Open','High','Low', 'Close','Volume',	'media5d')

k_best_features = SelectKBest(k= 'all')
k_best_features.fit_transform(features,label)
k_best_features_score = k_best_features.scores_
raw_pairs = zip(feature_list[1:], k_best_features_score)
ordered_pairs = list(reversed(sorted(raw_pairs, key= lambda x: x[1])))

k_best_features_final = dict(ordered_pairs[:15])
best_features = k_best_features_final.keys()
print('Melhores features:')
print(k_best_features_final)
```

```python
#Separando as melhores features
features = df_cotacoes.drop(['Low', 'media5d'], axis=1)
```

```python
#Normalizando os dados de entrada (features)
#Gerando o novo padrão
scaler = MinMaxScaler().fit(features)
features_scale = scaler.transform(features)

print('features:', features_scale.shape)
print(features_scale)



```

```python
#Separação dos dados treinamento, teste e validação
x_train = features_scale[:qtd_linha_treino]
x_test = features_scale[qtd_linha_treino:qtd_linha_teste]

y_train = label[:qtd_linha_treino]
y_test = label[qtd_linha_treino:qtd_linha_teste]
```

```python
#Treinamento com regressão linear
lr = linear_model.LinearRegression()
lr.fit(x_train, y_train)
pred = lr.predict(x_test)
cd= r2_score(y_test, pred)

f"Coeficiente de determinação :  {cd * 100: .2f}"
```

```python
#Executando a previsão
previsao = features_scale[qtd_linha_teste:qtd_linhas]

res_full = df_cotacoes['Adj Close']
res = res_full[qtd_linha_teste:qtd_linhas]

pred = lr.predict(previsao)

df = pd.DataFrame({'real':res, 'previsao': pred})
df['real'] = df['real'].shift(+1)

df
```
